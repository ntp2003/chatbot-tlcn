```python
import jsonlines
file_path_phone = "phone_data.jsonl"
file_path_laptop = "laptop_data.jsonl"
file_path_accessory = "accessories_data.jsonl"
```

```python
unique_phone_brands = set()
unique_laptop_brands = set()
unique_phone_product_types = set()
unique_laptop_product_types = set()
unique_accessory_brands = set()
unique_accessory_product_types = set()

with jsonlines.open(file_path_phone, mode='r') as reader:
    for item in reader:  # Đọc từng dòng trong file JSONL
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_phone_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_phone_product_types.add((product_type['code'], product_type['name']))

with jsonlines.open(file_path_laptop, mode='r') as reader:
    for item in reader: 
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_laptop_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_laptop_product_types.add((product_type['code'], product_type['name']))

with jsonlines.open(file_path_accessory, mode='r') as reader:
    for item in reader:  
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_accessory_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_accessory_product_types.add((product_type['code'], product_type['name']))

print("---Unique Phone Brands---:")
for code, name in unique_phone_brands:
    print(f"Code: {code}, Name: {name}")

print("\n---Unique Phone Product Types---:")
for code, name in unique_phone_product_types:
    print(f"Code: {code}, Name: {name}")

print("\n---Unique Laptop Brands---:")
for code, name in unique_laptop_brands:
    print(f"Code: {code}, Name: {name}")

print("\n---Unique Laptop Product Types---:")
for code, name in unique_laptop_product_types:
    print(f"Code: {code}, Name: {name}")

print("\n---Unique Accessory Brands---:")
for code, name in unique_accessory_brands:
    print(f"Code: {code}, Name: {name}")

print("\n---Unique Accessory Product Types---:")
for code, name in unique_accessory_product_types:
    print(f"Code: {code}, Name: {name}")
```

```python
import json
# Sử dụng set để lưu trữ unique brands và product types
unique_brands = set()
unique_product_types = set()

unique_brands_output_file_path = "unique_brands.json"
unique_product_types_output_file_path = "unique_product_types.json"

# Đọc data từ file phone_data.jsonl
with jsonlines.open(file_path_phone, mode='r') as reader:
    for item in reader:
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_product_types.add((product_type['code'], product_type['name']))

# Đọc data từ file laptop_data.jsonl
with jsonlines.open(file_path_laptop, mode='r') as reader:
    for item in reader:
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_product_types.add((product_type['code'], product_type['name']))

with jsonlines.open(file_path_accessory, mode='r') as reader:
    for item in reader:
        brand = item.get('brand')
        product_type = item.get('productType')
        if brand:
            unique_brands.add((brand['code'], brand['name']))
        if product_type:
            unique_product_types.add((product_type['code'], product_type['name']))

# In ra để kiểm tra
print("---All Unique Brands---:")
for code, name in unique_brands:
    print(f"Code: {code}, Name: {name}")

print("\n---All Unique Product Types---:")
for code, name in unique_product_types:
    print(f"Code: {code}, Name: {name}")



# Chuyển set thành list of dictionaries để lưu vào JSON
unique_brands_list = [{"code": code, "name": name} for code, name in unique_brands]
unique_product_types_list = [{"code": code, "name": name} for code, name in unique_product_types]

# Lưu vào file JSON
with open(unique_brands_output_file_path, 'w', encoding='utf-8') as outfile:
    json.dump(unique_brands_list, outfile, ensure_ascii=False, indent=4)

with open(unique_product_types_output_file_path, 'w', encoding='utf-8') as outfile:
    json.dump(unique_product_types_list, outfile, ensure_ascii=False, indent=4)


print(f"\nĐã lưu dữ liệu vào {unique_brands_output_file_path} và {unique_product_types_output_file_path}")
```

```python
unique_phone_product_status = set()
unique_phone_status_onweb = set()
unique_laptop_product_status = set()
unique_laptop_status_onweb = set()
unique_accessory_product_status = set()
unique_accessory_status_onweb = set()
with jsonlines.open(file_path_phone, mode='r') as reader:
    for item in reader:  # Đọc từng dòng trong file JSONL
        skus = item.get('skus')
        for sku in skus:
            product_status = sku.get('productStatus')
            status_onweb = sku.get('statusOnWeb')
            if product_status:
                unique_phone_product_status.add((product_status['code'], product_status['displayName']))
            if status_onweb:
                unique_phone_status_onweb.add((status_onweb['code'], status_onweb['displayName']))

with jsonlines.open(file_path_laptop, mode='r') as reader:
    for item in reader:  # Đọc từng dòng trong file JSONL
        skus = item.get('skus')
        for sku in skus:
            product_status = sku.get('productStatus')
            status_onweb = sku.get('statusOnWeb')
            if product_status:
                unique_laptop_product_status.add((product_status['code'], product_status['displayName']))
            if status_onweb:
                unique_laptop_status_onweb.add((status_onweb['code'], status_onweb['displayName']))

with jsonlines.open(file_path_accessory, mode='r') as reader:
    for item in reader:
        skus = item.get('skus')
        for sku in skus:
            product_status = sku.get('productStatus')
            status_onweb = sku.get('statusOnWeb')
            if product_status:
                unique_accessory_product_status.add((product_status['code'], product_status['displayName']))
            if status_onweb:
                unique_accessory_status_onweb.add((status_onweb['code'], status_onweb['displayName']))

print("---Phone product status---:")
for code, name in unique_phone_product_status:
    print(f"Code: {code}, Name: {name}")

print("\n---Phone status on web---:")
for code, name in unique_phone_status_onweb:
    print(f"Code: {code}, Name: {name}")

print("\n---Laptop product status---:")
for code, name in unique_laptop_product_status:
    print(f"Code: {code}, Name: {name}")

print("\n---Laptop status on web---:")
for code, name in unique_laptop_status_onweb:
    print(f"Code: {code}, Name: {name}")

print("---Accessory product status---:")
for code, name in unique_accessory_product_status:
    print(f"Code: {code}, Name: {name}")

print("\n---Accessory status on web---:")
for code, name in unique_accessory_status_onweb:
    print(f"Code: {code}, Name: {name}")
```
