```python
import jsonlines
file_path = "fine-tuning/fine-tuning_data/train_tone_ds1.jsonl"
```

```python
import json

# Open and read the .jsonl file
with open(file_path, "r", encoding="utf-8") as file:
    user_messages = []
    for line in file:
        data = json.loads(line)  # Parse each JSON line
        for msg in data.get("messages", []):  
            if msg.get("role") == "user":
                user_messages.append(msg.get("content"))


print(user_messages)
```

```python
from uuid import uuid4
```

```python

'''

import sys
import os
sys.path.append(os.path.abspath("..")) 
sys.path.append(os.path.abspath("../models"))  # Th√™m th∆∞ m·ª•c cha (root) v√†o sys.path
'''
from models.user import UserRole
from repositories.user import create as create_user, CreateUserModel
from repositories.thread import create as create_thread, CreateThreadModel
from service.store_chatbot_v2 import gen_answer
```

```python
def get_actual_answer(input: str) -> str:
    user = create_user(
        CreateUserModel(user_name=str(uuid4()), role=UserRole.chainlit_user)
    )
    thread = create_thread(CreateThreadModel(user_id=user.id, name=user.user_name))

    return gen_answer(
        thread_id=thread.id,
        history=[{"role": "user", "content": str(input)}],
        user_id=user.id,
    )
```

```python
"""Get all the answer from each input in user_messages"""
answers = []
for msg in user_messages:
    answers.append(get_actual_answer(msg))

print(answers)
```

```python
import pandas as pd

# Chuy·ªÉn danh s√°ch c√¢u tr·∫£ l·ªùi th√†nh DataFrame
df = pd.DataFrame({"User Message": user_messages, "Answer": answers})

pd.set_option("display.max_colwidth", None)
df.style.set_properties(**{'text-align': 'left'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'left')]}]
)

# Hi·ªÉn th·ªã b·∫£ng d·ªØ li·ªáu
df
```

```python
df.to_csv("dataframe_output.csv", index=False, encoding="utf-8-sig")
```

```python
import json

# T·∫°o danh s√°ch JSON t·ª´ d·ªØ li·ªáu
jsonl_data = [{"User Message": msg, "Answer": ans} for msg, ans in zip(user_messages, answers)]

# Xu·∫•t ra file JSONL
with open("answers.jsonl", "w", encoding="utf-8") as f:
    for item in jsonl_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print("write file answers.jsonl success")
```
